```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('../ENCFF780PGS.bedpe', sep='\t', header=None, names=['chr1', 'start1', 'end1', 'chr2', 'start2', 'end2', 'score'])
```

```python
df
```

```python
df = df[df['chr1'] == df['chr2']]
```

```python
df = df[df['chr1'] == 'chr10']
```

```python
df
```

```python
df = df[df['score']>10]
```

```python
df = df[(df['start1'] < df['start2']) & (df['end1'] < df['end2']) & (df['start1'] < df['end2'])]
```

```python
df['mid1'] = (df['start1'] + df['end1'])/2
df['mid2'] = (df['start2'] + df['end2'])/2
```

```python
scaler = MinMaxScaler(feature_range=(1, 1950))

df['mid1'] = scaler.fit_transform(df['mid1'].values.reshape(-1, 1))
df['mid2'] = scaler.fit_transform(df['mid2'].values.reshape(-1, 1))
```

```python
df['mid1'] = df['mid1'].astype(int)
df['mid2'] = df['mid2'].astype(int)
```

```python
df = df[abs(df['mid1'] - df['mid2']) > 10]
```
